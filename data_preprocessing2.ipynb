```python
import pandas as pd

# Read csv file
data = pd.read_csv('data/Features/systemic lupus_Dataset_Team_ML.csv')
labels = pd.read_csv('data/Labels/Systemic_Lupus_labels.txt', delimiter='\t')

print(data.head())
print(labels.head())
```

```python
# join dataset based on patient ids
# use left join to keep filtered data
data = pd.merge(data, labels[['UKB ID', 'time_window_incident/nan/10.0/nan/av1']], left_on='ID', right_on='UKB ID', how='left')
data = data.drop(columns=['UKB ID'])
```

```python
data.describe()
```

```python
# List of columns to drop
columns_to_drop = [
    'f.904.0.0_Do not know',
    'f.904.0.0_Prefer not to answer',
    'f.20116.0.0_Prefer not to answer',
    'f.6145_Prefer not to answer',
    'f.1558.0.0_Prefer not to answer'
]

# Drop the columns
data = data.drop(columns=columns_to_drop)
```

```python
len(data.columns)
```

```python
# Columns related to stress events
stress_columns = [
    'f.6145_Death of a close relative',
    'f.6145_Serious illness, injury or assault to yourself',
    'f.6145_Serious illness, injury or assault of a close relative',
    'f.6145_Financial difficulties',
    'f.6145_Marital separation/divorce',
    'f.6145_Death of a spouse or partner'
]

# Creating new column 'Experienced Stress'
data['Experienced Stress'] = data[stress_columns].any(axis=1).astype(int)

# Updating 'Experienced Stress' to 0 if 'f.6145_None of the above' is 1
data.loc[data['f.6145_None of the above'] == 1, 'Experienced Stress'] = 0

# Drop the original columns related to stress and 'f.6145_None of the above'
data = data.drop(columns=stress_columns + ['f.6145_None of the above'])
```

```python
# Create the new 'Gender' column
data['Gender'] = data['f.31.0.0_Female']

# Drop the original columns
data = data.drop(columns=['f.31.0.0_Female', 'f.31.0.0_Male'])
```

```python
data = data.drop(columns=['f.1558.0.0_Never', 'f.20116.0.0_Never'])
```

```python
column_renames = {
    'f.21022.0.0': 'Age at Recruitment',
    'f.21001.0.0': 'BMI',
    'f.904.0.0': 'Number of days/week of vigorous physical activity 10+ minutes',
    'f.20107_Father_Bowel cancer': 'Father with Bowel Cancer',
    'f.20107_Father_Diabetes': 'Father with Diabetes',
    'f.20110_Mother_Bowel cancer': 'Mother with Bowel Cancer',
    'f.20110_Mother_Diabetes': 'Mother with Diabetes',
    'f.20116.0.0_Current': 'Current Smoker',
    'f.20116.0.0_Previous': 'Previous Smoker',
    'f.6177_Blood pressure medication': 'Blood Pressure Medication',
    'f.6177_Cholesterol lowering medication': 'Cholesterol Lowering Medication',
    'f.1558.0.0_Daily or almost daily': 'Alcohol intake frequency: Daily',
    'f.1558.0.0_Once or twice a week': 'Alcohol intake frequency: Once or twice a week',
    'f.1558.0.0_One to three times a month': 'Alcohol intake frequency: One to three times a month',
    'f.1558.0.0_Special occasions only': 'Alcohol intake frequency: Special occasions only',
    'f.1558.0.0_Three or four times a week': 'Alcohol intake frequency: Three or four times a week',
    'f.6177.0.0_Insulin': 'Insulin',
}

# rename columns
data = data.rename(columns=column_renames)
```
